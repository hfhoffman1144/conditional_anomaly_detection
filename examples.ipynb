```python
DATA_PATH = 'data/accepted_2007_to_2018Q4.csv'
```

```python
import numpy as np
import polars as pl
import matplotlib.pyplot as plt
pl.Config().set_tbl_cols(200)
#pd.set_option('display.max.columns', None)
```

```python
data = pl.scan_csv(DATA_PATH, ignore_errors=True)
data.head().collect()
```

```python
q = (
    data
    .groupby("fico_range_low")
    .agg(
        [   
            pl.mean("int_rate"),
            pl.count(),
            
        ]
    )
)
avg_rate_by_fico = q.collect()
avg_rate_by_fico
```

```python
plt.scatter(avg_rate_by_fico['fico_range_low'], avg_rate_by_fico['int_rate'])
```
